```python
import pandas as pd
df = pd.read_csv('心血管100健康資料練習檔_學生.csv',encoding='utf-8')
display(df)
```

```python
height=df['身高'] 
weight=df['體重'] 
bmi = weight/(height/100)**2

underweight=0
health=0
overweight=0
fat=0
for i in range(len(bmi)):
    if bmi[i]<18.4:
        underweight+=1
    elif bmi[i]<23.9:
        health+=1
    elif bmi[i]<27.9:
        overweight+=1
    else:
        fat+=1
print("過輕的人有%d位"%underweight)
print("健康的人有%d位"%health)
print("過重的人有%d位"%overweight)
print("肥胖的人有%d位"%fat)
```

```python
import matplotlib.pyplot as plt
```

```python
total=[underweight,health,overweight,fat]
category = ["underweight", "health", "overweight", "fat"]


plt.figure(figsize=(12,8))# 設定圓餅圖大小
separeted = (0, 0, 0, 0)# 依據類別數量，分別設定要突出的距離      


plt.pie(total,                          # 數值
        labels = category,              # 標籤
        autopct = "%1.1f%%",            # 將數值百分比並留到小數點一位
        explode = separeted,            # 設定分隔的區塊位置
        pctdistance = 0.5,              # 數字距圓心的距離
        textprops = {"fontsize" : 12},  # 文字大小
        shadow=False)                   # 設定陰影

 
plt.axis('equal')                                          # 使圓餅圖比例相等
plt.title("Pie chart of BMI", {"fontsize" : 18})           # 設定標題及其文字大小
plt.legend(loc = "best")                                   # 設定圖例及其位置為最佳
plt.show()      
```

```python
from ipywidgets import interact, interact_manual
```

```python
def BMI(身高,體重):
    height = float(身高)
    weight = float(體重)
    bmi = weight/(height/100)**2
    print("你的 BMI = %.2f" % bmi)
    if bmi<18.4:
        print('過輕!')
    elif bmi<23.9:
        print('健康!')
    elif bmi<27.9:
         print('過重!')
    else:
         print('肥胖')
```

```python
interact_manual(BMI, 身高="請輸入你的身高", 體重="請輸入你的體重")
```
